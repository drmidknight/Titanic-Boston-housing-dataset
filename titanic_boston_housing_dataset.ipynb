```python
# Fill missing values for 'Age" with the median, and 'Embacked' with the most frequent value
classification_data['Age'].fillna(classification_data['Age'].median(), inplace=True)
```

```python
# Detecting and Handling Outliners
from scipy import stats
```

```python
# Selecting numerical columns for outliner handling
numerical_columns = ['Age', 'Fare']
```

```python
# Z-Score outliner detection
z_scores = np.abs(stats.zscore(classification_data[numerical_columns]))
```

```python
# Filtering out rows with z-scores > {potential outliners}
classification_data = classification_data[(z_scores < 3).all(axis=1)]
```

```python
classification_data
```

```python
# Checking for Influential Data Points
#We can check for influential data points using statistical measures like Cook's distance or leverage in more advanced regression models, but in this case, we’ll focus on detecting extreme outliers using z-scores.
```

```python
# Checking Normality and Features - PLotting histogram to check for normality visually
classification_data[numerical_columns].hist(bins=20, figsize=(10, 5))
plt.show()
```

```python
#Data Transformation
#You might transform skewed features using log transformations. For the Titanic dataset, we'll apply transformations on the Fare feature, which is known to be highly skewed.
```

```python
# Apply log transformation to 'Fare' to reduce skewness
classification_data['Fare'] = np.log1p(classification_data['Fare'])
```

```python
# Feature Selection
# Use domain knowledge or feature importance techniques (e.g., correlation, decision trees) to drop irrelevant features.
```

```python
#Dropping irrelevant features (eg. Passenger ID, Name , Ticket)
classification_data.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
classification_data
```

```python
# Convert categorical variables to numerical using one-hot encoding
classification_data = pd.get_dummies(classification_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
#  Detect and Handle Outliers
#For regression, we'll use the Interquartile Range (IQR) method to detect and remove outliers.
```

```python
# Calculate Q1 (25TH PERCENTILE) AND Q3 (75THY PERCENTILE)
Q1 = regression_data.quantile(0.25)
Q3 = regression_data.quantile(0.75)
IQR = Q3 - Q1
```

```python
# Removing Outliners
regression_data = regression_data[~((regression_data < (Q1 - 1.5 * IQR)) | (regression_data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Checking for Influential data points
# We can use leverage statistics and Cook’s distance to identify influential data points in more advanced regression models, but for simplicity, we’ll focus on removing outliers.
```
